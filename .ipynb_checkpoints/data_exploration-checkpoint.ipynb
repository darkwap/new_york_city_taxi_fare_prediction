### 1. load data

```python
#load all data and subset
import numpy as np
import pandas as pd

#subset 500k because of analyze/calculation time.
data = pd.read_csv('data/train.csv',nrows=500000)

print(data.shape)
#drop row where column value are missing
data.dropna(inplace = True)
print( data.shape)
```

```python
#describe all data
data.describe()
```

### 2. new york city coordinates

```python
# new york coordinates square
#longitude, latitude
#all new york area
LEFT_LONGITUDE = -74.3
RIGHT_LONGITUDE= -73.7
TOP_LATITUDE = 40.9
BOTTOM_LATITUDE = 40.5
data.drop(data[(data['pickup_longitude'] > RIGHT_LONGITUDE) | (data['pickup_longitude'] < LEFT_LONGITUDE) | (data['pickup_longitude'] == 0)].index, inplace = True)
data.drop(data[(data['pickup_latitude'] < BOTTOM_LATITUDE) | (data['pickup_latitude'] > TOP_LATITUDE) | (data['pickup_latitude'] == 0)].index, inplace = True)
data.drop(data[(data['dropoff_longitude'] > RIGHT_LONGITUDE) | (data['dropoff_longitude'] < LEFT_LONGITUDE) | (data['dropoff_longitude'] == 0)].index, inplace = True)
data.drop(data[(data['dropoff_latitude'] < BOTTOM_LATITUDE) | (data['dropoff_latitude'] > TOP_LATITUDE) | (data['dropoff_latitude'] == 0)].index, inplace = True)
data.drop(data[(data['pickup_latitude'] == data['dropoff_latitude']) & (data['pickup_longitude'] == data['dropoff_longitude'])].index, inplace = True)
print(data.shape)
```

### 3. describe data

```python
#describe our subset
data.describe()
```

### 4. remove unnecesary data

```python
#fare_amount < 0 it is impossible so lets remove them
data = data[data.fare_amount>=0]
#data for passenger count = 0 also is useless for me
data = data[data.passenger_count > 0]
```

### 5. plot graph

```python
# This function is based on https://stackoverflow.com/questions/27928/
# distance between two points using haversine formula
def distance(lat1, lon1, lat2, lon2):
    p = 0.017453292519943295 # Pi/180
    a = 0.5 - np.cos((lat2 - lat1) * p)/2 + np.cos(lat1 * p) * np.cos(lat2 * p) * (1 - np.cos((lon2 - lon1) * p)) / 2
    return 12742 * np.arcsin(np.sqrt(a))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
%matplotlib inline
```

```python
#histogram numbers of rides vs fare
# plot histogram of fare
data[data.fare_amount<100].fare_amount.hist(bins=100,figsize=(14,3) )
plt.xlabel('fare $USD')
plt.title('Histogram');
```

```python
data_distance = pd.DataFrame()
data_distance['distance'] = distance(data.pickup_latitude, data.pickup_longitude, \
                                      data.dropoff_latitude, data.dropoff_longitude)

data_distance.distance.hist(bins=50, figsize=(12,4))
plt.xlabel('distance km')
plt.title('Histogram ride distances in km')
```

```python
#add distance to data
data['distance'] = data_distance
```

### 6. plot data points on map

```python
#ny city map with pickup, dropoff points
# thanks to:  https://www.kaggle.com/breemen/nyc-taxi-fare-data-exploration
BB = (LEFT_LONGITUDE, RIGHT_LONGITUDE, BOTTOM_LATITUDE, TOP_LATITUDE)
nyc_map = plt.imread('data/nyc_-74.3_-73.7_40.50_40_90.PNG')
```

```python
# this function will be used more often to plot data on the NYC map
# thanks to: https://www.kaggle.com/breemen/nyc-taxi-fare-data-exploration
def plot_on_map(df, BB, nyc_map, s=10, alpha=0.2):
    fig, axs = plt.subplots(1, 2, figsize=(16,10))
    axs[0].scatter(df.pickup_longitude, df.pickup_latitude, zorder=1, alpha=alpha, c='r', s=s)
    axs[0].set_xlim((BB[0], BB[1]))
    axs[0].set_ylim((BB[2], BB[3]))
    axs[0].set_title('Pickup locations')
    axs[0].imshow(nyc_map, zorder=0, extent=BB)

    axs[1].scatter(df.dropoff_longitude, df.dropoff_latitude, zorder=1, alpha=alpha, c='r', s=s)
    axs[1].set_xlim((BB[0], BB[1]))
    axs[1].set_ylim((BB[2], BB[3]))
    axs[1].set_title('Dropoff locations')
    axs[1].imshow(nyc_map, zorder=0, extent=BB)
```

```python
plot_on_map(data, BB, nyc_map, s=1, alpha=0.3)
```

```python
#timestamp provided in data set is not good for analyze
# we should split this for years, days, hours etc

data['pickup_datetime'] = data['pickup_datetime'].str.slice(0, 16)
data['pickup_datetime'] = pd.to_datetime(data['pickup_datetime'], utc=True, format='%Y-%m-%d %H:%M')

data['pickup_hour'] = data['pickup_datetime'].dt.hour
data['pickup_day'] = data['pickup_datetime'].dt.day
data['pickup_month'] = data['pickup_datetime'].dt.month
data['pickup_year'] = data['pickup_datetime'].dt.year
```

```python
#delta latitude, longitude, from pickup to dropoff
data['delta_longitude'] = data['pickup_longitude'] - data['dropoff_longitude']
data['delta_latitude'] = data['pickup_latitude'] - data['dropoff_latitude']
```
